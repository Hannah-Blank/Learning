```python
FN = "Darius-13-100m-Fly.txt"
FOLDER = "swimdata/"
with open(FOLDER + FN) as df:
    data= df.readlines()
times = data[0].strip().split(",")
```

```python
split_fn= FN.removesuffix(".txt").split("-")
swimmer, age_group, distance, stroke = split_fn
print(swimmer, age_group, distance, stroke)
```

```python
converts = []
for time in times:
    minutes, rest = time.split(":")
    seconds, hundredths = rest.split(".")
    converted_time = int(minutes)*60*100 + int(seconds)*100 + int(hundredths)
    converts.append(converted_time)
    print(time, "->", converted_time)

average_time = sum(converts) / len(converts)
print("Average time:", average_time)
```

```python
import statistics
average =statistics.mean(converts)
```

```python
hundredths = round(int(average_time % 100), 2)
seconds = round(int((average_time // 100) % 60), 2)
minutes = round(int(average_time // (60*100)), 2)
print(f"Average time: {minutes}:{seconds}.{hundredths}")
```
